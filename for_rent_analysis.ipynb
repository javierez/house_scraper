```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

#### Tratamiento de DataFrame Alquileres

```python
df=pd.read_csv('for_rent.csv')
df
```

```python
# Remove the columns with null in título and links
df = df[df['Título'].notnull()]
df = df[df['Links'].notnull()]
# Convert "Precio" column to float
df["Precio"] = df["Precio"].str.replace(" € /mes", "")
df["Precio"] = df["Precio"].str.replace('.','').astype(int)
#Sustituir por algún regex que identifique cualquier tipo de letra o whitespace
df["Planta"] = df["Planta"].replace("\n", None)
df["Planta"] = df["Planta"].replace(" ", None)
```

```python
df['Planta'].astype(float).mean()
```

```python
average_floor=math.ceil(df['Planta'].astype(float).mean())
df['Planta']=df['Planta'].fillna(average_floor)
```

```python
average_n_bath=math.ceil(df['Baños'].astype(float).mean())
df['Baños']=df['Baños'].fillna(average_n_bath)
```
